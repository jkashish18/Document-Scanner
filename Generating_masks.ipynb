```python
import cv2
import os
import numpy as np
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
annot_path = os.getcwd()+'\\Annotations'
path = os.getcwd()+'\\photos'
for filename in os.listdir(path):
    img = cv2.imread(os.path.join(path,filename))
    x = []
    y = []
    height, width = img.shape[:2]
    black = np.zeros((height, width))
    file = filename.replace('jpg', 'xml', 3)
    tree = ET.parse(os.path.join(annot_path,file))
    print(file)
    for elem in tree.iter(tag='pt'):
        x.append(elem[0].text)
        y.append(elem[1].text)
    for i in range(4):
        black = cv2.line(black, (int(x[i]), int(y[i])), (int(x[(i+1) % 4]), int(y[(i+1) % 4])), (255,255,255), 4, cv2.LINE_4)
    cv2.imwrite(os.getcwd()+'\\Maskedimages\\'+filename, black)
    plt.imshow(black)
    plt.show()
```
